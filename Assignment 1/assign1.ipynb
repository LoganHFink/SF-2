5. The data in `data_df` contains noisy data.  In this step you are to clean the noisy data (or extract the clean data). Your cleaned data **for each run**
- must not include the data that is after once the filter hits the motion sensor
- must include only the bit where a drop in position starts.  You should not include any constant position values before the drop of positions takes place.    
Note that you should clean your data for each of the 10 runs and you should automate this process in your code with logical conditions, loops, etc.
6. Store your clean data into a new DataFrame called `clean_df`. 
7. Note that now that the values in the **Time** columns may have shifted due to the clean-up.  Update your `clean_df` to rescale all the **Time** columns to start the time from 0.0 in increments of 0.02. 
8. Write this to a new .csv file called `"clean_data_multiple_filter.csv"`. 

```python
# Your code for steps 5-8 go here:

clean_df = pd.DataFrame([])

df_list = []

for i in range(1,11):
    current_df = pd.DataFrame([])
    current_df = data_df[[f'Time (s) Run #{i}',f'Position (m) Run #{i}']]

    done = False
    current_end = 0

    while not done:
        current_num = current_df[f'Position (m) Run #{i}'].iloc[current_end]
        current_end += 1
        if current_num < 0.14:
            done = True

    value = current_df.loc[:,f'Position (m) Run #{i}'].iloc[0:current_end]

    current_df.loc[:,f'Position (m) Run #{i}'] = value

    # Here increments for time are actually 0.02 since the frequency on the sensor was correct for these ten runs.     

    done = False
    current_index = len(current_df)

    while not done:
        nums = list(current_df[f'Position (m) Run #{i}'].iloc[current_index-5:current_index])
        if nums[0] - nums[4] == 0:
            start = current_index-4
            done = True
        current_index -= 1

    current_df = current_df[start-1:].dropna()

    # Sets times to start from zero

    # Make lists, then combine into clean_df to avoid errors of different length columns. Make all lengths of 70 by adding None up to desired length

    new_times = []
    index_column = current_df[f'Time (s) Run #{i}']
    for time in index_column:
        new_times.append(float(round(time - index_column.iloc[0],2)))

    while len(new_times) < 70:
        new_times.append(None)

    pos_list = []
    for pos in current_df[f'Position (m) Run #{i}']:
        pos_list.append(pos)

    while len(pos_list) < 70:
        pos_list.append(None)

    clean_df[f'Position (m) Run #{i}'] = pos_list
    clean_df[f'Time (s) Run #{i}'] = new_times

clean_df.to_csv('clean_data_multiple_filter.csv')
```
