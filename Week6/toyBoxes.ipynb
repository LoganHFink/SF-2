# Toy Boxes:

We are given n unopened boxes with k number of figurines in each box. The boxes cant be opened , the order of the figurines cant be changed. A box cant be rotated (otherwise the figurines will be facing the wrong way)

Each figurine has a specified height, i.e: in a given box the height of figurines from left to right could be: 4, 5, 7.

Note that the number of figurines per box may vary, but nonempty.

We would like to organise all the boxes s.t they are aranged in increasing order of heights form left to right.

Not all inputs are possible, write a program to determine if we have a such arrangement or not

# Input Specification:

- first line: integer n representing number of boxes

- next n lines: one for each box, each line begins with the following:

--> integer k figurines in each box (k>=1)

--> k integers indicating height of figurines, seperated by a space (heights >= 1)

# Example Input:

2    
<!-- 2 is n -->
3 4 5 7             
<!-- 3 is k, [4,5,7] are heights -->
2 6 8               
<!-- 2 is k, [6,8] are heights -->
Output should be NO

2                   
<!-- 2 is n -->
3 4 5 7
<!-- 3 is k, [4,5,7] are heights -->
2 8 9               
<!-- 2 is k, [8,9] are heights -->
Output should be YES

# Top Down Design:

- Capturing main tasks of the solution, some tasks might take little code that can be solved directly, and others are big and will require functions.

```python
def readBoxes(n):
    boxes = []
    for _ in range(n):
        box = list(input(f'k, heights: ').split())
        box = [int(h) for h in box[1:]]
        boxes += [box]
    return boxes

def boxOk(box):
    return sorted(box) == box

def allBoxesOk(boxes):
    # determine if each box in boxes has figurines in increasing order of height. Returns True or False
    ok = True
    for box in boxes:
        if not boxOk(box):
            ok = False
    return ok

def boxIntervals(boxes):
    # returns just the first and last h per box
    output = []
    for box in boxes:
        output += [[box[0],box[-1]]]
    return output

def intervalsOk(intervals):
    intervals.sort()
    prev = intervals[0][1]
    ordered = True
    for i in intervals[1:]:
        if i[0] < prev:
            ordered = False
        prev = i[1]
    return ordered

# Start of Program:

n = int(input(f'n: '))
boxes = readBoxes(n)
if not allBoxesOk(boxes):
    print("NO")
else:   
    intervals = boxIntervals(boxes)
    if intervalsOk(intervals):
        print("YES")
    else:
        print("NOOOO")
```
