```python
standings = {
 "Bruins": [28, 28, 8],
 "Canadiens": [30, 26 ,6],
 "Penguins": [24, 30, 10],
 "Predators": [23, 32, 7],
 "Jets": [43, 16, 4],
 "Oilers": [36, 22, 4] }

def orderedTeams(standings):
    output = []
    for team in standings:
        points = 2*standings[team][0] + standings[team][1]
        output.append([team,points])
    print(output)
    
orderedTeams(standings)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = {'Months': ['Jan','Feb','Mar'],
      'Elec': [50,30,25],
      'Clothing': [40,28,73],
      'Furniture': [20,100,9]}

df = pd.DataFrame(df)

mean_sales = []
for category in df.keys():
    if category != 'Months':
        mean = (df[category]).mean()
        mean_sales.append((float(mean),category))
print(mean_sales)

legend = []

for category in df.keys():
    if category != 'Months':
        x = df['Months']
        y = df[category]
        plt.plot(x,y)
        legend.append(category)

plt.legend(legend)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Title')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
 'Math': [85, 92, 78, 88, 95],
 'Science': [90, 87, 82, 84, 91],
 'English': [88, 91, 79, 85, 92]}
df = pd.DataFrame(data) 

averages = []

for i in range(len(df)):
    temp = df.iloc[i,1:]
    # print(temp.mean())
    averages.append(temp.mean())

df['Average'] = averages

df

for i in range(len(df)):
    x = df['Student'].iloc[i]
    y = df['Average'].iloc[i]
    plt.bar(x,y)

plt.xlabel('Students')
plt.ylabel('Averages')
plt.title('title')
```

```python
t = ((3, 2, 6), (5, 2, 1), (4, 8))
def elementsTuple(t: tuple):
    l = []
    for tup in t:
        for i in range(len(tup)):
            l.append(tup[i])
    l = tuple(sorted(l))
    return l
print(elementsTuple(t)) 
```

```python
lst = [[5, 5, 4],
 [8, 4, 7],
 [2, 5, 1]]
def checkSum(lst):
    sum1 = 0
    sum2 = 0
    for i in range(len(lst)):
        sum1 += lst[i][i]
        sum2 += lst[i][-1-i]
    return sum1 == sum2
print(checkSum(lst)) 
```

```python
lst_o_dict = [ {'destination': 'USA', 'flights': 11},
{'destination': 'France', 'flights': 19},
{'destination': 'USA', 'flights': 20},
{'destination': 'France', 'flights': 10},
{'destination': 'Morocco', 'flights': 27},
{'destination': 'China', 'flights': 35},
{'destination': 'Morocco', 'flights': 9} ] 

def flights_data(lst_o_dict):
    output = {}

    for x in lst_o_dict:
        destination,flights = x['destination'],x['flights']
        if destination not in output.keys():
            output[destination] = 0
        output[destination] += flights
    return output
        
flights_data(lst_o_dict)
```

```python
def transform_tuple(data):
    a, (b, c) = data
    b = list(b)
    for i in range(len(b)):
        b[i] += c[i]
    new_tuple = (sum(b) * a, tuple(b), c[::-1])
    return new_tuple
data = (3, ((4, 6, 8), (5, 7, 9)))
result = transform_tuple(data)
print(data)
print(result) 
```
